```python
df = pd.read_csv("Levels_Fyi_Salary_Data.csv")
df = df[df.basesalary != 0]
df[['city', 'state']] = df['location'].str.split(',', 1, expand=True)
df['correct_state'] = df['state'].apply(lambda x: len(x)<=3)
df = df[df.correct_state == True]
tags = ['Machine Learning', 'Security', 'Full Stack','API Development','Web Development','QA-Testing','Site Reliability','DevOps','Networking','Distributed Systems']
df['tag'] = np.random.choice(tags, len(df))
df['title'].unique()
titles = ['Software Engineering Manager','Software Engineer', 'Data Scientist']
df = df.loc[df['title'].isin(titles)]
df['yearsofexperience'] = df['yearsofexperience'].apply(lambda x: int(x))
df['basesalary'] = df['basesalary'].apply(lambda x: int(x))
df['stockgrantvalue'] = df['stockgrantvalue'].apply(lambda x: int(x))
df['bonus'] = df['bonus'].apply(lambda x: int(x))
df['level'] = df.level.str.extract('(\d+)')
df['levelcorrect'] = df['level'].apply(lambda x: (not x) or float(x)<=8)
df['level'] = df['level'].apply(lambda x: x if x else random.randint(1,7))
df = df[df.levelcorrect == True]
df = df.sample(1987)
df_post = df[['timestamp', 'company', 'title', 'tag', 'level', 'basesalary', 'stockgrantvalue', 'bonus', 'otherdetails', 'city', 'state']]
df_post.head(20)
```

```python
fakedlevel = df[['company', 'title']].groupby(by=["company", "title"]).first()
fakedlevel['system'] = np.random.choice(["IC", "L", "E"], size=len(fakedlevel))
```

```python
df_user = df[['yearsofexperience', 'Education']]
df_user['Education'] = pd.Series(random.choices(['BS', 'MS', 'BA', 'MA', 'PhD', 'HS'], weights=[5, 3, 5, 1, 1, 1], k=len(df_user)), index=df_user.index)
df_user
```

```python
fake = faker.Faker()
np.random.seed(np.random.randint(10))
fake.seed_instance(np.random.randint(10))

def get_fake():
    gender = np.random.choice(["M", "F"], p=[0.5, 0.5])
    first_name = fake.first_name_male() if gender=="M" else fake.first_name_female()
    last_name = fake.last_name()
    domain = np.random.choice(["gmail.com", "yahoo.com"], p=[0.5, 0.5])
    email= f"{first_name}.{last_name}@{domain}"
    return gender, first_name, last_name, email

df_user[["gender", "first_name", "last_name", "email"]] = df_user.apply(lambda row: get_fake(), axis='columns', result_type='expand')
df_user = df_user.drop_duplicates('email')
df_user = df_user.sample(467)
```
